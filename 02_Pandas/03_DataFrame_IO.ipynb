## Pandas Data I/O
CSV, JSON, HTML, SAS, Python Pickling 등 다양한 포맷의 데이터 파일을 읽어 DataFrame을 생성할 수 있다.

```python
import pandas as pd
```

### writefile 명령으로 csv 파일 생성

```python
%%writefile ./sample1.csv
c1, c2, c3
1, 1.11, one
2, 2.22, two
3, 3.33, three
```

### CSV 파일로부터 데이터 읽기

```python
pd.read_csv('./sample1.csv')
```

```python
%%writefile ./sample2.csv
1, 1.11, one
2, 2.22, two
3, 3.33, three
```

```python
pd.read_csv('./sample2.csv', names=['c1', 'c2', 'c3'])
```

#### 특정 column을 인덱스로 지정할 수도 있다.

```python
pd.read_csv('./sample1.csv', index_col='c1')
```

```python
%%writefile ./sample3.txt
c1        c2        c3        c4 
0.179181 -1.538472  1.347553  0.43381
1.024209  0.087307 -1.281997  0.49265
0.417899 -2.002308  0.255245 -1.10515
```

#### read_table() 또한 외부 데이터를 읽을 때 사용할 수 있고 sep 인수를 사용해 구분자를 따로 설정할 수 있다.

```python
pd.read_table('./sample3.txt', sep='\s+')
```

```python
%%writefile ./sample4.txt
파일 제목: sample4.txt
데이터 포맷의 설명: 
c1, c2, c3
1, 1.11, one
2, 2.22, two
3, 3.33, three
```

```python
pd.read_csv('./sample4.txt', skiprows=[0, 1])
```

```python
%%writefile ./sample5.csv
c1, c2, c3
1, 1.11, one
2,, two
누락, 3.33, three
```

```python
df = pd.read_csv('./sample5.csv', na_values=['누락'])
df
```

### DataFrame 값을 CSV 파일로 출력

```python
data = {
    "2015": [9904312, 3448737, 2890451, 2466052],
    "2010": [9631482, 3393191, 2632035, 2431774],
    "2005": [9762546, 3512547, 2517680, 2456016],
    "2000": [9853972, 3655437, 2466338, 2473990],
    "지역": ["수도권", "경상권", "수도권", "경상권"],
    "2010-2015 증가율": [0.0283, 0.0163, 0.0982, 0.0141]
}
columns = ["지역", "2015", "2010", "2005", "2000", "2010-2015 증가율"]
index = ["서울", "부산", "인천", "대구"]
df = pd.DataFrame(data, index=index, columns=columns)
df
```

```python
df.to_csv('./sample6.csv')
```

```python
df.to_csv('sample7.csv', index=False, header=False)
```
